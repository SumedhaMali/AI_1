Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
#Task3
#Identify outlier countries in South Asia based on Score and GDP

Q1_score=south_asia_df['score'].quantile(0.25)
Q3_score=south_asia_df['score'].quantile(0.75)
IQR_score=Q3_score-Q1_score


Q1_gdp=south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp=south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp=Q3_gdp-Q1_gdp

upper_bound_score=Q3_score+1.5*IQR_score
lower_bound_score=Q1_score-1.5*IQR_score
upper_bound_gdp=Q3_gdp+1.5*IQR_gdp
lower_bound_gdp=Q1_gdp-1.5*IQR_gdp

#identifying outliers
outliers=south_asia_df[
    (south_asia_df['score']>upper_bound_score) | (south_asia_df['score']<lower_bound_score) |
    (south_asia_df['Log GDP per capita']>upper_bound_gdp) | (south_asia_df['Log GDP per capita']<lower_bound_gdp)
]

outliers[['Country name','score','Log GDP per capita']]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], label='Countries', color='blue')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], label='Outliers', color='red', marker='x')
plt.axhline(y=upper_bound_score, color='green', linestyle='--', label='Upper Bound')
plt.axhline(y=lower_bound_score, color='black', linestyle='--', label='Lower Bound')
plt.axvline(x=upper_bound_gdp, color='green', linestyle='--')
plt.axvline(x=lower_bound_gdp, color='black', linestyle='--')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.title('Outlier Detection: GDP per Capita vs Score')
plt.legend()
plt.show()

# Step 4: Print the outliers
print("Outliers based on Score or GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

Outliers can have a disproportionate impact on regional averages, potentially distorting the perception of overall economic health, social support, and life expectancy. It's important to identify and consider these outliers when analyzing regional data to make sure policies or conclusions are based on a representative sample of the entire region.

6. Outlier Detection:
• Identify outlier countries in both regions based on Score and GDP per Capita.
• Plot these outliers and discuss their implications.

```python
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

south_asia_score_outliers = find_outliers_iqr(south_asia_df, 'score')
south_asia_gdp_outliers = find_outliers_iqr(south_asia_df, 'Log GDP per capita')


middle_east_score_outliers = find_outliers_iqr(middle_east_countries_df, 'score')
middle_east_gdp_outliers = find_outliers_iqr(middle_east_countries_df, 'Log GDP per capita')

print("Outlier Countries in South Asia based on Score:")
print(south_asia_score_outliers[['Country name', 'score']])
print("\nOutlier Countries in South Asia based on Log GDP per capita:")
print(south_asia_gdp_outliers[['Country name', 'Log GDP per capita']])
print("\nOutlier Countries in Middle East based on Score:")
print(middle_east_score_outliers[['Country name', 'score']])
print("\nOutlier Countries in Middle East based on Log GDP per capita:")
print(middle_east_gdp_outliers[['Country name', 'Log GDP per capita']])
```

```python
south_asia_all_outliers = pd.concat([south_asia_score_outliers, south_asia_gdp_outliers])

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, color='blue', label='South Asia Countries')
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_all_outliers, color='red', label='Outliers')
plt.title("Outlier Countries in South Asia")
plt.xlabel('Log GDP per capita')
plt.ylabel('Happiness Score')
plt.legend()
plt.tight_layout()
plt.show()


middle_east_all_outliers = pd.concat([middle_east_score_outliers, middle_east_gdp_outliers])

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_countries_df, color='blue', label='Middle East Countries')
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_all_outliers, color='red', label='Outliers')
plt.title("Outlier Countries in Middle East")
plt.xlabel('Log GDP per capita')
plt.ylabel('Happiness Score')
plt.legend()
plt.tight_layout()
plt.show()
```
